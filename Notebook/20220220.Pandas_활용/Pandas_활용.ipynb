## data type 보기
가격 등이 object 로 읽어져서 number 로 읽어야 할 때 등

```python
df.dtypes
```

```python
# df['분양가격(제곱미터)'] 가 1,000 처럼 comma 를 가지고 있어서, to_numeric() 사 NaN 이 나옴.
# 명시적으로, object => string > float 64 로 변환해야 함.
df['분양가격(제곱미터)'] = df['분양가격(제곱미터)'].astype(str)
df.tail()
```

```python
# df['분양가격(제곱미터)'] 가 1,000 처럼 comma 를 가지고 있어서, to_numeric() 사 NaN 이 나옴.
# 명시적으로, object => string > float 64 로 변환해야 함.
df['분양가격(제곱미터)'] = df['분양가격(제곱미터)'].apply(lambda x: x.replace(',',''))
df.tail()
```

```python
# df['분양가격(제곱미터)'] = pd.to_numeric(df['분양가격(제곱미터)'])
# ==> ValueError: Unable to parse string "  " at position 28

df['분양가격(제곱미터)'] = pd.to_numeric(df['분양가격(제곱미터)'], errors='coerce')
df.dtypes

# 이 부분에서 6,000 이럴 때는 NaN 으로 변환되어버림.
```

## 그룹화 (Grouping)

```python
df.head()
```

```python
df.shape
```

```python
df.groupby(['지역명','연도','월'])['분양가격(제곱미터)'].agg('sum')
```

```python
df.groupby(['지역명','연도','월']).count()
```

```python
# df.groupby(['지역명']).count() # ==> 동일함.
df.groupby('지역명').count()
```

```python
# pandas 의 groupby 를 이용.count() 를 구하여 중복된 값 찾기
df_cnt = df.groupby(['지역명','연도','월']).count()
df_cnt
# df_cnt.to_csv('df_cnt.csv')
```
