<데이터셋의 변수들>

Country: 모두 알다시피 나라입니다... 해당 기관에선 총 149개국을 조사했네요. (row가 149개겠죠?)

Regional_indicator: 세계의 대표적인 지역 10개로 묶어놓은 범주형 변수입니다 (동아시아, 서유럽 등등...)

Ladder_score: 각 국가의 국민들에게 전반적인 행복지수에 대한 설문을 진행해서 나온 점수입니다.
(0~10점, 높을 수록 행복함)

Logged_GDP_per_capita: 모두가 한번쯤 들어보셨을 GDP입니다! 다른 변수들과 scale 차이가 크다보니 해당 데이터에선 로그를 씌워서 최대 11,12 정도가 되게 해놨네요.

Social_support: '문제가 생겼을 때 언제든지 도와줄 수 있는 사람이 있나요?' 에 대한 답을 Yes는 1, No는 0으로 하여 얻은 국가별 평균값입니다.

Healthy_life_expectancy: 각 나라의 기대 수명입니다.

Freedom_to make_life_choices: '본인이 하려던 것에 주어진 자유에 만족하나요?'에 대한 답변을 통해 얻은 평균입니다.

Generosity: '지난 달에 기부를 한적이 있나요?'에 대한 대답과 GDP를 회귀 분석하여 나온 residual 값입니다.

Pereception_of_corruption: '정부에 부정, 부패가 만연한가요?', '직장에 부정, 부패가 만연한가요?' 에 대한 답변을 통해 얻은 평균입니다.

```python
df2 = df.copy()
df2
```

```python
df2.columns[:2]
```

```python
df2.rename(columns = {'Country name': 'country_name'}, inplace=True)
```

```python
df2.rename(columns = {'Regional indicator': 'region'}, inplace=True)
df2.columns[:2]
```

```python
df2.eval('my_loc = country_name + "__" + region')
# 음. 역시 두개의 문자를 합치는 것은 힘들군.
```

```python
df2['my_loc'] = df2.apply(lambda row: row['country_name'] + '__' + row['region'], axis=1)
```

```python
df2.loc[:,['my_loc','country_name','region']]
```
